```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV data
df = pd.read_csv('test_dataset.csv')

# Create a directed graph using NetworkX
G = nx.DiGraph()

# Add nodes and attributes (comment_id as node ID and sarcasm_score as node attribute)
for index, row in df.iterrows():
    G.add_node(row['comment_id'], sarcasm=row['sarcasm_score'])

# Add edges based on parent-child relationship (parent_id and comment_id)
for index, row in df.iterrows():
    if row['parent_id'] != 'NA':  # Ensure the comment has a parent
        G.add_edge(row['parent_id'], row['comment_id'])

# Define the layout for the graph (spring layout is often useful for visualizing graphs)
pos = nx.spring_layout(G)

# Define node colors based on sarcasm score, with a default value for nodes without the sarcasm attribute
node_color = [
    'red' if G.nodes[node].get('sarcasm', 0) == 1 else 'blue'  # Default to 0 if 'sarcasm' attribute is missing
    for node in G.nodes
]

# Plot the graph
plt.figure(figsize=(12, 8))

# Draw the nodes
nx.draw_networkx_nodes(G, pos, node_color=node_color, node_size=500, alpha=0.8)

# Draw the edges
nx.draw_networkx_edges(G, pos, alpha=0.5)

# Add labels (optional, can be removed if too cluttered)
nx.draw_networkx_labels(G, pos, font_size=10)

# Display the plot
plt.title('Graph of Reddit Comments with Sarcasm Labels')
plt.show()
```

```python
import networkx as nx
import pandas as pd
import torch
from torch_geometric.utils import from_networkx
import numpy as np

# Load the CSV data
df = pd.read_csv('test_dataset.csv')

# Create a directed graph using NetworkX
G = nx.DiGraph()

# Add nodes and attributes (comment_id as node ID and sarcasm_score as node attribute)
for index, row in df.iterrows():
    G.add_node(row['comment_id'], sarcasm=row['sarcasm_score'], feature=[row['comment_score']])

# Add edges based on parent-child relationship (parent_id and comment_id)
for index, row in df.iterrows():
    if row['parent_id'] != 'NA':  # Ensure the comment has a parent
        G.add_edge(row['parent_id'], row['comment_id'])

# Ensure all nodes have the same attributes (sarcasm and feature)
for node in G.nodes:
    if 'sarcasm' not in G.nodes[node]:
        G.nodes[node]['sarcasm'] = 0  # Default to non-sarcastic (0)
    if 'feature' not in G.nodes[node]:
        G.nodes[node]['feature'] = [0]  # Default to a feature of 0

# Convert to PyTorch Geometric format
data = from_networkx(G)

# Assign node features and labels
x = torch.tensor([node['feature'] for node in G.nodes.values()], dtype=torch.float)
y = torch.tensor([node['sarcasm'] for node in G.nodes.values()], dtype=torch.long)

data.x = x
data.y = y

# Create a train_mask
num_nodes = data.num_nodes
indices = np.arange(num_nodes)
np.random.shuffle(indices)

# Define the split: e.g., 80% training and 20% testing
train_size = int(0.8 * num_nodes)
train_indices = indices[:train_size]
test_indices = indices[train_size:]

# Create masks
train_mask = torch.zeros(num_nodes, dtype=torch.bool)
train_mask[train_indices] = True
test_mask = torch.zeros(num_nodes, dtype=torch.bool)
test_mask[test_indices] = True

data.train_mask = train_mask  # Add the train_mask to the data object
data.test_mask = test_mask
```

```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv

class GCN(torch.nn.Module):
    def __init__(self, num_node_features, hidden_channels, num_classes):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(num_node_features, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, num_classes)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index

        # Perform first graph convolution
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, training=self.training)
        
        # Perform second graph convolution
        x = self.conv2(x, edge_index)

        return F.log_softmax(x, dim=1)

# Instantiate the model
model = GCN(num_node_features=data.num_node_features, hidden_channels=16, num_classes=2)

# Define optimizer and loss function
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
loss_fn = torch.nn.CrossEntropyLoss()

# Training loop (simplified)
def train():
    model.train()
    optimizer.zero_grad()
    out = model(data)
    loss = loss_fn(out[data.train_mask], data.y[data.train_mask])
    loss.backward()
    optimizer.step()

for epoch in range(200):
    train()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Generate positions for visualization
pos = nx.spring_layout(G)

# Get predictions
model.eval()
out = model(data)  # Get predictions
pred = out[data.test_mask].max(1)[1]  # Get the predicted classes

# Get test node indices (as actual node IDs)
test_indices = data.test_mask.nonzero(as_tuple=True)[0]  # Indices of test nodes

# Prepare colors for test node visualization
node_color = []
valid_nodelist = []  # To store valid nodes for visualization

for i in test_indices:
    actual_index = i.item()  # Convert tensor to item for indexing
    node_id = data.x[actual_index].item()  # Ensure this retrieves actual node ID
    if node_id in G:  # Check if the node exists in the graph
        valid_nodelist.append(node_id)
        # Get the predicted class for this node
        predicted_class = pred[test_indices == actual_index].item()
        actual_class = data.y[actual_index].item()
        node_color.append('green' if predicted_class == actual_class else 'red')

# Create a subgraph of valid nodes
subgraph = G.subgraph(valid_nodelist)
pos = nx.spring_layout(subgraph)  # Recalculate positions for valid nodes

# Draw the subgraph
plt.figure(figsize=(12, 8))
nx.draw(subgraph, pos, node_color=node_color, with_labels=True, node_size=500)
plt.title('GNN Predictions Visualization')
plt.show()
```
